#### 1.Univariate Selection 

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df['price_range']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
#### Applay SelectKbest Algorithm
order_rank_feature = SelectKBest(score_func=chi2,k=20)
order_feature=order_rank_feature.fit(x,y)
```

```python
order_feature.scores_
```

```python
df_score=pd.DataFrame(order_feature.scores_,columns=["Score"])
dfcolumns = pd.DataFrame(x.columns)
```

```python
feature_rank = pd.concat([dfcolumns,df_score],axis=1)
```

```python
feature_rank.columns=['Feature','Score']
```

```python
feature_rank
```

```python
feature_rank.nlargest(10,'Score')
```

#### 2.Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
```

```python
model = ExtraTreesClassifier()
```

```python
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
ranked_feature=pd.Series(model.feature_importances_,index=x.columns)
ranked_feature
```

```python
ranked_feature.nlargest(10).plot(kind='barh')
```

```python
#### Remove the corralated
```

```python
threshold = 0.8
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
correlation(df.iloc[:,:-1],threshold)
```
